## View data

```python
np.random.seed(42)
data = ImageDataBunch.from_folder(path, train=".", valid_pct=0.2,
         ds_tfms=get_transforms(), size=256, num_workers=0).normalize(imagenet_stats)
```

```python
data.show_batch(rows=3,figsize=(10,10))
```

```python
learn = create_cnn(data, models.resnet34, metrics=error_rate)
```

```python
learn.fit_one_cycle(4)
#this one gives ~ 72% accuracy
```

```python
learn.save('stage-1')
```

```python
learn.unfreeze()
```

```python
#finding better learning rates
learn.lr_find()
```

```python
learn.recorder.plot()
```

```python
learn.fit_one_cycle(10, max_lr=slice(1e-5,1e-3))
#this one gives ~ 87% accuracy
```

```python
learn.save('stage-2')
```

## Interpretation

```python
interp = ClassificationInterpretation.from_learner(learn)
```

```python
interp.plot_confusion_matrix()
```

So model has some difficulties distinguish between Puma, Adidas and Nike. Timberland and Lacoste are ok.

## Ploting top losses

```python
losses,idxs = interp.top_losses()

len(data.valid_ds)==len(losses)==len(idxs)
```
